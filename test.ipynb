```python
import smtplib, ssl
import os
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.image import MIMEImage

sender_email = 'govindplaystation4028@gmail.com'
recipients = ['govindsomadas@gmail.com']
password = 'nvka suug mepj evpp'  # Use App Password

# Images folder
image_dir = "images"

# Automatically get all image files (jpg, jpeg, png, gif)
image_files = [f for f in os.listdir(image_dir)
               if f.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))]

# Load HTML template
with open("./email-templates/index.html", "r", encoding="utf-8") as f:
    html_template = f.read()

# Build <img> tags for all images
images_html = "".join([f'<img src="cid:{i}"><br>' for i in image_files])
html_content = html_template.replace("{{images}}", images_html)
print(html_content)

# Create main multipart/related container
msg = MIMEMultipart("related")
msg['Subject'] = "TestEmail_012"
msg['From'] = sender_email
msg['To'] = ", ".join(recipients)

# Create alternative part (plain text + HTML)
alt_part = MIMEMultipart("alternative")
alt_part.attach(MIMEText("This is the plain text fallback.", "plain"))
alt_part.attach(MIMEText(html_content, "html"))

msg.attach(alt_part)

# Attach each image to the related part
for i, image_file in enumerate(image_files):
    image_path = os.path.join(image_dir, image_file)
    with open(image_path, "rb") as f:
        img = MIMEImage(f.read())
        img.add_header('Content-ID', f"<{i}>")
        img.add_header('Content-Disposition', 'inline', filename=image_file)
        msg.attach(img)

# Send the email
context = ssl.create_default_context()
with smtplib.SMTP("smtp.gmail.com", 587) as server:
    server.starttls(context=context)
    server.login(sender_email, password)
    server.sendmail(sender_email, recipients, msg.as_string())

print(f"Email sent with {len(image_files)} inline images.")
```
